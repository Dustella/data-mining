```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn import preprocessing

data_raw = pd.read_csv("../data/bankloan.csv")

# col A16 is y
X, y = data_raw.iloc[:, :-1], data_raw.iloc[:, -1]
```

```python
# use label encoder to encode categorical data
le = preprocessing.LabelEncoder()

X = X.apply(le.fit_transform)
X
```

```python

# split data into train and test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, )
```

```python
# import and use RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)

clf.fit(X_train, y_train)
```

```python
# predict and use accuracy_score, confusion_matrix

from sklearn.metrics import accuracy_score 
from sklearn.metrics import confusion_matrix

y_pred = clf.predict(X_test)

acc = accuracy_score(y_test, y_pred)

print("Accuracy: ", acc)


cm = confusion_matrix(y_test, y_pred)

cm
```
